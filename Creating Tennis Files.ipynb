```python
from sqlalchemy import create_engine
import pandas as pd

engine = create_engine('postgresql://madelinevossbrinck:localhost@localhost:5432/tennis')

aus_men = pd.read_csv('./tennis/AusOpen-men-2013.csv')

aus_men.to_sql('aus_men', engine, index=False)
```

```python
query = 'SELECT * FROM aus_men;'
df = pd.read_sql(query, engine)

df.head()
```

```python
aus_women = pd.read_csv('./tennis/AusOpen-women-2013.csv')

aus_women.to_sql('aus_women', engine, index=False)
query = 'SELECT * FROM aus_women;'
df = pd.read_sql(query, engine)

df.head()
```
