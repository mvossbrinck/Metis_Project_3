### Investigate Missing Containment Data

```python
# Read in data set
fires = pd.read_csv('Fires_During_After_2000.csv')
```

```python
# Create variable for when there's no containment time information
fires['NO_CONT'] = np.where(fires['CONT_DATE'].isna(), 'Missing', 'Not Missing')
fires['NO_CONT'].value_counts()
```

```python
# Check missing % by source system type 
data_crosstab = pd.crosstab(fires['SOURCE_SYSTEM_TYPE'], 
                            fires['NO_CONT'],  
                               margins = False, normalize = 'index') 
print(data_crosstab) 
```

```python
# Check missing by source system type 
data_crosstab = pd.crosstab(fires['SOURCE_SYSTEM_TYPE'], 
                            fires['NO_CONT'],  
                               margins = False) 
print(data_crosstab) 
```

I thought that those records with containment times could be a sign that the data in that record is more complete and reliable. The ones from the federal source system almost always had containment information, so I do think the model could be reasonably applied to those. ### EDA for Final Data Set

```python
# Read in data source
fires = pd.read_csv('Fires_Final_Data.csv')
```

```python
# Create binary target
fires['FIRE_SIZE_CLASS_BIN'] = np.where(fires['FIRE_SIZE']> 10, 1, 0)
fires['FIRE_SIZE_CLASS_BIN'].value_counts()
```

```python
# Correlation matrix
fires.corr()
```

```python
# Correlation heatmap
fig, ax = plt.subplots(figsize=(12,12)) 
sns.heatmap(fires.corr(), cmap="seismic", annot=True, vmin=-1, vmax=1, ax=ax);

# No strong correlations with target, some expected high correlations between related 
# variables i.e. Latitude and Longitude
```

```python
sns.pairplot(fires);
```

```python
# Table of fire cause by fire size class bin (small = 0, large = 1)
data_crosstab = pd.crosstab(fires['STAT_CAUSE_DESCR'], 
                            fires['FIRE_SIZE_CLASS_BIN'],  
                               margins = False, normalize = 'index') 
print(data_crosstab) 

# There is a difference between the different fire causes, although no one cause jumps out
# as extremely different
```

```python
# Box plot of fire size by fire cause
sns.set_theme(style="whitegrid")
sns.boxplot(x = 'FIRE_SIZE', y = 'STAT_CAUSE_DESCR', data = fires);

# The actual boxplots don't show up because of the numerous large outliers
# I tried them with size restrictions and it still didn't work well so didn't use 
# boxplots for any other size chart
```

```python
# Table of region by fire size class bin (small = 0, large = 1)
data_crosstab = pd.crosstab(fires['REGION'], 
                            fires['FIRE_SIZE_CLASS_BIN'],  
                               margins = False, normalize = 'index') 
print(data_crosstab) 

# Northeast is way less likely to have large fires - the final XGBoost model gives high
# importance to the Northeast feature 
```

```python
# Scatter of discovery date by fire size
plt.scatter(data=fires, x='DISCOVERY_DATE_DT', y='FIRE_SIZE', marker='o', s=25);

# There seems to be a wavelike trend, possibly seasonal
```

```python
# Scatter of discovery day of year by fire size
plt.scatter(data=fires, x='DISCOVERY_DOY', y='FIRE_SIZE', marker='o', s=25);

# Fires appear to get the biggest in early spring, summer, and early fall
```
